```python
def map_category(venue_df, user_id):
    testset = pd.read_csv(user_id, sep="\t", 
                   header=None,  encoding = "ISO-8859-1", 
                   names=['user_id', 'venue_id', 'time_utc', 'time_offset'])
    df = pd.DataFrame(columns=['user_id', 'venue_id', 'latitude', 'longitude', 
                            'time_utc', 'time_offset', 'category_name', 'country_code'])
    for index, row in testset.iterrows():
        x=((venue_df[venue_df['venue_id']==row['venue_id']]).values)[0]
        user_id=row['user_id']
        venue_id=row['venue_id']
        lat=x[1]
        lng=x[2]
        time=row['time_utc']
        offset=row['time_offset']
        catname=x[3]
        cc=x[4]
        print(index)
        df = df.append([{'user_id': user_id, 'venue_id':venue_id, 'latitude':lat, 'longitude':lng, 
                        'time_utc':time, 'time_offset':offset, 'category_name':catname, 
                         'country_code':cc}], ignore_index=True)
    return df
        
```
